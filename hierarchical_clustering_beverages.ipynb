# Hierarchical Clustering Beverages

Иерархическая кластеризация данных о напитках.

Файл показывает:
- как построить матрицу слияния через linkage;
- как визуализировать дендрограмму;
- как "разрезать" дендрограмму на кластеры;
- как посмотреть статистику по найденным кластерам.

```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, fcluster, linkage
```

## 1. Базовые настройки и загрузка данных

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
beverages_df = pd.read_csv('data/beverage_ratings.csv', sep=';', index_col='numb.obs')
```

```python
print(beverages_df)
```

## 2. Построение иерархической кластеризации

linkage:
- принимает матрицу признаков;
- считает расстояния между объектами;
- по шагам объединяет похожие объекты в кластеры.

ward       -> метод объединения кластеров;

euclidean  -> метрика расстояния.

```python
linkage_matrix = linkage(beverages_df, 'ward', 'euclidean')
```

```python
print(linkage_matrix)
```

## 3. Дендрограмма

Дендрограмма показывает весь процесс слияния:
- листья — исходные объекты;
- ветви — объединения кластеров;
- высота ветви — расстояние между объединяемыми группами.

```python
dendrogram(linkage_matrix, orientation='right')
plt.title('Дендрограмма напитков')
plt.show()
```

## 4. Разрез дендрограммы на кластеры

fcluster присваивает каждому объекту номер кластера.

criterion='distance':

объединения продолжаются, пока расстояние не превысит заданный порог.

Здесь порог = 3.

```python
beverages_df['cluster'] = fcluster(linkage_matrix, 3, criterion='distance')
```

## 5. Интерпретация кластеров

Средние значения признаков в каждом кластере.

```python
print(beverages_df.groupby('cluster').mean())
```

Подробная статистика по кластерам.

```python
print(beverages_df.groupby('cluster').describe())
```

Количество объектов в каждом кластере.

```python
print(beverages_df.groupby('cluster').size())
```
