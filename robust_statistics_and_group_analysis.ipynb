## 2. Пример группового анализа на Ames Housing

```python
# Загружаем данные о ценах на дома.
housing_df = pd.read_csv('data/ames_housing.txt', sep='\t', header=0, index_col=False)

# Группируем по MS Zoning.
zoning_groups = housing_df.groupby('MS Zoning')

# Сводная статистика по группам.
# print(zoning_groups.describe())

# Гистограммы SalePrice по типам зонирования.
zoning_groups['SalePrice'].plot.hist(alpha=0.6)
plt.legend()
plt.show()
```

```python
# Boxplot по SalePrice и MS Zoning.
# Удобен для сравнения медианы, размаха и выбросов.
housing_df.boxplot(column='SalePrice', by='MS Zoning')
plt.title('SalePrice по зонам')
plt.suptitle('')
plt.show()

# Частоты категорий в MS Zoning.
print(housing_df['MS Zoning'].value_counts())
```

## 3. Почему среднее плохо работает при выбросах

```python
# Данные о населении городов России до 1959 года.
towns_df = pd.read_csv('data/russian_towns_1959.csv', header=0, encoding='cp1251', index_col=0)

# Сколько процентов городов имеют население ниже среднего.
print(len(towns_df[towns_df['население'] < 52.925199]) / len(towns_df) * 100)

# Убираем Москву и Санкт-Петербург как крайние наблюдения.
towns_without_outliers_df = towns_df.iloc[2:1004, :]

# Проверяем тот же расчет после удаления двух крупнейших городов.
print(
    len(towns_without_outliers_df[towns_without_outliers_df['население'] < 44.997904]) /
    len(towns_without_outliers_df) * 100
)

# Вывод:
# даже после удаления крупных выбросов среднее остается плохим кандидатом
# на роль "типичного" наблюдения.

# Медиана устойчивее к выбросам.
print(towns_df['население'].median())

# Гистограмма исходного распределения.
towns_df['население'].hist(bins=60)
plt.show()

# Гистограмма после логарифмирования.
np.log(towns_df['население']).hist(bins=60)
plt.show()
```

## 4. Визуальный анализ швейцарских банкнот

Этот блок активен по умолчанию.

Здесь задача — посмотреть, насколько хорошо genuine / counterfeit
разделяются визуально по нескольким признакам.

```python
banknotes_df = pd.read_csv('data/swiss_bank_notes.csv', index_col=0)
```

Красим два класса в разные цвета.

```python
banknote_colors = (
    banknotes_df['Status']
    .replace('genuine', 'green')
    .replace('counterfeit', 'red')
)
```

scatter_matrix — матрица попарных диаграмм рассеяния.

Ее удобно использовать, чтобы увидеть,
по каким комбинациям признаков классы разделяются лучше всего.

```python
scatter_matrix(
    banknotes_df,
    figsize=(6, 6),
    diagonal='kde',
    c=banknote_colors,
    alpha=0.2,
)
plt.show()
```

Отдельный scatter-график по двум признакам.

```python
banknotes_df.plot.scatter(x='Top', y='Bottom', c=banknote_colors)
plt.show()
```
