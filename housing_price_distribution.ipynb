# Housing Price Distribution

Анализ распределения цен на дома (Ames Housing).

Файл показывает базовые шаги исследовательского анализа:
- загрузка данных;
- общая статистика по датасету;
- визуальный анализ распределения;
- логарифмирование;
- сглаженная оценка плотности.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde
```

## 1. Загрузка данных

```python
housing_df = pd.read_csv('data/ames_housing.txt', sep='\t', header=0, index_col=False)
```

## 2. Базовая сводка по датасету

```python
print(housing_df.shape)                    # количество строк и столбцов
print(len(housing_df))                     # только количество строк
print(housing_df.dtypes)                   # типы данных по столбцам
print(housing_df.info())                   # общая техническая информация о датасете
print(housing_df.describe(include='all'))  # статистика по числовым и категориальным полям
```

## 3. Гистограммы по SalePrice

```python
# Простой вариант гистограммы.
plt.style.use('seaborn-v0_8-deep')

plt.hist(housing_df['SalePrice'], bins=60, density=True)
plt.title('Распределение цен на дома')
plt.xlabel('SalePrice')
plt.ylabel('Плотность')
plt.show()

# Логарифмирование помогает, если распределение сильно скошено вправо.
plt.hist(np.log(housing_df['SalePrice']), bins=200, density=True)
plt.title('Логарифмированное распределение SalePrice')
plt.xlabel('log(SalePrice)')
plt.ylabel('Плотность')
plt.show()
```

## 4. KDE + гистограмма

```python
# gaussian_kde строит сглаженную оценку плотности распределения.
price_density = gaussian_kde(housing_df['SalePrice'], bw_method=0.1)

# Сетка значений по оси X от минимальной до максимальной цены.
x_grid = np.linspace(housing_df['SalePrice'].min(), housing_df['SalePrice'].max(), 1000)

# Линия плотности.
plt.plot(x_grid, price_density(x_grid), 'g')

# Полупрозрачная гистограмма поверх той же метрики.
plt.hist(housing_df['SalePrice'], density=True, alpha=0.3)

plt.title('Гистограмма и KDE для цен домов')
plt.xlabel('Цена дома')
plt.ylabel('Плотность')
plt.show()
```
